# PCA Part

```python
import numpy
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

### features variable contains the columns included in PCA

```python
#benin.columns
features = ['water_source', 'toilet', 'Electricity', 'Radio', 'Television','Fridge', 'Bicycle', 'Motorbike', 'Car', 'floor']
info = ['HHID', 'year']
```

```python
X = benin.loc[:, features].values
Y = benin.loc[:,info].values
```

```python
X = preprocessing.StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components = 1)
```

```python
X_pca = pca.fit_transform(X)
```

```python
print(pca.explained_variance_ratio_)
```

```python
principalDf = pd.DataFrame(data = X_pca, columns = ['wealth_index'])
```

### Creating a dataframe with the newly calculated wealth index

```python
final_benin = pd.concat([principalDf, benin[['HHID']], benin[['cluster_number']], benin[['household_number']], benin[['Year']]], axis = 1)
final_benin
```

```python
final_benin.dropna(inplace=True)
```

```python
final_benin.isna().sum() 
```

```python
final_benin.to_excel("Benin_18_wealth.xlsx")
```

### This creates a file that has wealth index grouped by cluster. Just what we want

```python
wealth_index_2018 = final_benin.groupby('cluster_number')['wealth_index'].mean()
```
